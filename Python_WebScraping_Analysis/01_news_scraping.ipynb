### 1-1) 필수과제 - Daum 뉴스기사 제목 스크래핑 하기

```python
# IT/과학 뉴스 
req_param = {
    'sid' : 'tech',
}

url = 'https://news.daum.net/news/{sid}'.format(**req_param)
print(url)



req_header = {
'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/135.0.0.0 Safari/537.36 Edg/135.0.0.0'
}


res = requests.get(url, headers=req_header)
res.encoding = 'utf-8'

print(res)
print(type(res))
if res.ok:
    html = res.text
    soup = BeautifulSoup(html, 'html.parser')
    a_tag_list = soup.select("ul.list_newsheadline2 > li > a[href]")
    print(type(a_tag_list))
    for a_tag in a_tag_list:
        title = a_tag.text.strip()
        link = a_tag['href']
        print(link,'\n', title, '\n')
else:
    print(f"에러코드 = {res.status_code}")

```

##### 1-2) 필수 과제
여러개의 section 중 하나를 선택해
url에서 뉴스기사의 링크와 제목을 출력하는 코드를 함수로 작성


```python
import requests
from bs4 import BeautifulSoup

section_dict = {
    "기후/환경": "climate",
    "사회": "society",
    "경제": "economy",
    "정치": "politics",
    "국재": "world",
    "문화": "culture",
    "생활": "life",
    "IT/과학": "it-tech",
    "인물": "people",
}


def print_news(sid):
    section = section_dict[sid]
    req_param = {"sid": sid}

    url = "https://news.daum.net/news/{sid}".format(**req_param)
    print(f"=====> {url} {section} 뉴스 <=====")

    req_header = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/135.0.0.0 Safari/537.36 Edg/135.0.0.0"
    }

    res = requests.get(url, headers=req_header)
    res.encoding = "utf-8"

    if res.ok:
        html = res.text
        soup = BeautifulSoup(html, "html.parser")
        a_tag_list = soup.select("ul.list_newsheadline2 > li > a[href]")
        for a_tag in a_tag_list:
            title = a_tag.text.strip()
            link = a_tag["href"]
            print(link, "\n", title, "\n")
    else:
        print(f"에러코드 = {res.status_code}")

print_news('경제')
```

```python
print_news('사회')
```

```python
print_news('인물')
```

### 추가과제) Nate 뉴스기사 제목 스크래핑하기

```python
import requests
from bs4 import BeautifulSoup

# 섹션 딕셔너리
section_dict = {
    100: "최신뉴스",
    200: "정치",
    300: "경제",
    400: "사회",
    500: "세계",
    600: "IT/과학"
}

# 각 섹션의 뉴스를 출력하는 함수
def print_news():
    for sid, section_name in section_dict.items():
        url = f"https://news.nate.com/section?mid=n0{sid}"
        print(f"\n=====> {section_name} 뉴스 <=====")
        print(f"URL: {url}")

        req_header = {
            "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/135.0.0.0 Safari/537.36 Edg/"
        }

        res = requests.get(url, headers=req_header)
        res.encoding = "euc-kr"

        if res.ok:
            soup = BeautifulSoup(res.text, "html.parser")
            a_tag_list = soup.select("ul.mduStrongList li h2 a[href]")

            if not a_tag_list:
                print("뉴스를 찾을 수 없습니다.\n")
                continue

            for a_tag in a_tag_list:
                title = a_tag.text.strip()
                link = a_tag["href"]
                if link.startswith("//"):
                    link = "https:" + link
                print(f"{title}\n{link}\n")
        else:
            print(f"에러코드 = {res.status_code}")

# 함수 실행
print_news()
```

### Nate 뉴스기사 제목 스크래핑하기

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
from IPython.display import Image, display

# 섹션 딕셔너리
section_dict = {
    100: "최신뉴스",
    200: "정치",
    300: "경제",
    400: "사회",
    500: "세계",
    600: "IT/과학"
}

def print_img_news():
    for sid, section_name in section_dict.items():
        base_url = f"https://news.nate.com/section?mid=n0{sid}"
        print(f"\n=====> {section_name} 뉴스 <=====")
        print(f"URL: {base_url}")

        headers = {
            "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/135.0.0.0 Safari/537.36"
        }

        response = requests.get(base_url, headers=headers)
        response.encoding = "euc-kr"

        if response.ok:
            soup = BeautifulSoup(response.text, "html.parser")

            title_tag = soup.select_one("div.mlt01 a span.tb h2.tit")
            link_tag = soup.select_one("div.mlt01 a")
            img_tag = soup.select_one("div.mlt01 a span.ib img")


            if title_tag:
                title = title_tag.text.strip()
                print(f"제목: {title}")
            else:
                print("(제목 없음)")


            if link_tag and link_tag.has_attr("href"):
                link = link_tag["href"]
                if link.startswith("//"):
                    link = "https:" + link
                print(f"링크: {link}")

            if img_tag and img_tag.has_attr("src"):
                img_src = img_tag["src"]
                full_img_url = urljoin("https:", img_src)
                print("대표 이미지:")
                display(Image(url=full_img_url))
            else:
                print("(이미지 없음)\n")
        else:
            print(f"에러코드: {response.status_code}")


print_img_news()
```

### 하나의 네이버 웹툰과 1개의 회차에 대한 Image 다운로드 하기

```python
import os
import requests
from bs4 import BeautifulSoup

def download_one_episode(title, no, url):
    res = requests.get(url)
    if not res.ok:
        print(f'Error Code = {res.status_code}')
        return

    soup = BeautifulSoup(res.text, 'html.parser')

    img_tags = soup.select("img[src*='IMAG01']")
    img_url_list = [img_tag['src'] for img_tag in img_tags]


    folder_name = f"{title}_{no}화"
    if not os.path.isdir(folder_name):
        os.mkdir(folder_name)

    req_header = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/135.0.0.0 Safari/537.36"
    }

    for img_url in img_url_list:
        img_res = requests.get(img_url, headers=req_header)
        if img_res.ok:
            img_data = img_res.content
            file_path = os.path.join(folder_name, os.path.basename(img_url))
            with open(file_path, 'wb') as f:
                f.write(img_data)
        else:
            print(f"이미지 다운로드 실패: {img_url}")

download_one_episode(
    '낢이사는이야기', 
    48, 
    'https://comic.naver.com/webtoon/detail?titleId=833255&no=49&week=tue'
)
```
