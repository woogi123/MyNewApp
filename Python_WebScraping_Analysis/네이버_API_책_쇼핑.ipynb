### 네이버 API 책 가져오기
- 질문 1 해결 (with open ~)
- 질문 2 해결 (def display_book ~)
- 질문 3 해결 (book.html)
- 질문 4 해결 (def high_discount)
- 질문 6 해결 (def infinity_books)

```python
import requests
import os
import json
import pandas as pd
from dotenv import load_dotenv

load_dotenv()

def search_book(query):
    client_id = os.getenv("NAVER_CLIENT_ID")
    client_secret = os.getenv("NAVER_CLIENT_SECRET")

    headers = {
        'X-Naver-Client-Id': client_id,
        'X-Naver-Client-Secret': client_secret,
    }

    payload = {
        'query':  query,
        'display': 50,
        'sort': 'sim'
    }

    url = 'https://openapi.naver.com/v1/search/book.json'
    res = requests.get(url, params=payload, headers=headers)
    items_data = res.json()['items']

    items_list = []
    for item in items_data:
        book = {
            'title': item['title'],
            'link': item['link'],
            'image': item['image'],
            'author': item['author'],
            'discount': item['discount'],
            'publisher': item['publisher'],
            'pubdate': item['pubdate'],
            'isbn': item['isbn'],
            'description': item['description'].replace('\n', ' ').replace('\r', '')
        }
        items_list.append(book)

    os.makedirs('data', exist_ok=True)


    with open('data/books.json', 'w', encoding='utf-8') as file:
        json.dump(items_list, file, ensure_ascii=False, indent=4)
    print("books.json 저장")

    df = pd.DataFrame(items_list)
    df.to_html('data/books.html', index=False, escape=False)
    print("books.html 저장")


def display_books():
    with open('data/books.json', 'r', encoding='utf-8') as file:
        data = json.load(file)

    df = pd.DataFrame(data)

search_book("파이썬")


def high_discount(query):
    search_book(query)
    with open('data/books.json', 'r', encoding='utf-8') as file:
        data = json.load(file)

    df = pd.DataFrame(data)
    df['discount'] = df['discount'].str.replace(',', '').astype(int)
    high_discount_books = df[df['discount'] > 20000].sort_values(by='discount', ascending=False)

    high_discount_books = high_discount_books.reset_index(drop=True)


    print(high_discount_books[['title', 'author', 'discount', 'publisher', 'pubdate']])
        
    high_discount_books.to_html('data/high_discount_books.html', index=False, escape=False)

    
high_discount('파이썬')

def infinity_books(query):
    search_book(query)
    with open('data/books.json', 'r', encoding='utf-8') as file:
        data = json.load(file)

    df = pd.DataFrame(data)

    infinityBooks = df[df['publisher'] == '인피니티북스'].sort_values(by='discount', ascending=False)

    infinityBooks = infinityBooks.reset_index(drop=True)

    print(infinityBooks[['title', 'link', 'author', 'discount', 'publisher', 'pubdate', 'isbn']])
        
    infinityBooks.to_html('data/infinity_books.html', index=False, escape=False)

infinity_books('파이썬')


# // html 연습문제처럼 꾸미기 + 질문 3부터 이어서 + 3-2 하기
```

### 네이버 쇼핑 API 가져오기

```python
import requests
import os
import json
import pandas as pd
from dotenv import load_dotenv

load_dotenv()

def search_shops(query):
    client_id = os.getenv("NAVER_CLIENT_ID")
    client_secret = os.getenv("NAVER_CLIENT_SECRET")

    headers = {
        'X-Naver-Client-Id': client_id,
        'X-Naver-Client-Secret': client_secret,
    }

    payload = {
        'query':  query, #'가디건',
        'display': 50,
        'sort': 'sim'
    }


    url = 'https://openapi.naver.com/v1/search/shop.json'
    res = requests.get(url, params=payload, headers=headers)
    items_data = res.json()['items']

    items_list = []
    for item in items_data:
        shops = {
            'title': item['title'],
            'link': item['link'],
            'image': item['image'],
            'lprice': item['lprice'],
            'hprice': item['hprice'],
            'mallName': item['mallName'],
            'productId': item['productId'],
            'productType': item['productType'],
            'brand': item['brand'],
            'maker': item['maker'],
            'category1': item['category1'],
            'category2': item['category2'],
            'category3': item['category3'],
            'category4': item['category4'],
        }
        items_list.append(shops)

    os.makedirs('shops', exist_ok=True)


    with open('data/shops.json', 'w', encoding='utf-8') as file:
        json.dump(items_list, file, ensure_ascii=False, indent=4)
    print("shops.json 저장")

    df = pd.DataFrame(items_list)
    df.to_html('data/shops.html', index=False, escape=False)
    print("shops.html 저장")


def display_shops():
    with open('data/shops.json', 'r', encoding='utf-8') as file:
        data = json.load(file)

    df = pd.DataFrame(data)
    print(df)


search_shops("가디건")
display_shops()

def high_discount(query):
    search_shops(query)
    with open('data/shops.json', 'r', encoding='utf-8') as file:
        data = json.load(file)

    df = pd.DataFrame(data)
    df['lprice'] = df['lprice'].str.replace(',', '').astype(int)
    high_discount_shops = df[df['lprice'] > 50000].sort_values(by='lprice', ascending=False)

    high_discount_shops = high_discount_shops.reset_index(drop=True)


    print(high_discount_shops[['brand', 'lprice', 'mallName', 'link']])
        
    high_discount_shops.to_html('data/high_discount_shops.html', index=False, escape=False)

    
high_discount('가디건')

def naver_shop(query):
    search_shops(query)
    with open('data/shops.json', 'r', encoding='utf-8') as file:
        data = json.load(file)

    df = pd.DataFrame(data)

    naverShop = df[df['mallName'] == '네이버'].sort_values(by='lprice', ascending=True)

    naverShop = naverShop.reset_index(drop=True)

    print(naverShop[['lprice', 'hprice', 'mallName', 'productId', 'productType', 'brand']])
        
    naverShop.to_html('data/infinity_books.html', index=False, escape=False)

naver_shop('가디건')


# // html 연습문제처럼 꾸미기 + 질문 3부터 이어서 + 3-2 하기
```
